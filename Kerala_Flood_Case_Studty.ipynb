Q1
Question Explanation:

Let A represents : Flood
B represents: heavy rain in June

We need to calculate  P(A|B)  i.e.  P(A∩B)P(B)



```python
pd.crosstab(data['JUN'],
            data['FLOODS'],
            margins=True,
            margins_name='Total')
```

Now,  P(A∩B)  = Probability of Flood occuring AND heavy rainfall in JUNE

As we know in the contingency table, FLOODS = YES represents that flood has occured and JUN = 1 means heavy rainfall.

We need to check value where FLOODS = YES and JUN = 1 which is 44

Then by the formula of condititonal probability we can feed this data

```python
# probability of high rainfall in June P(J)
# P(J) = possible outcomes in june having heavy rainfall / total outcomes

P_J = (16+44)/(42+16+16+44)

# now, P(A and B) (Flood = YES and Jun = 1)

P_F_and_J = 44/(42+16+16+44)

#, so our probability of flood occurring given that the high rainfall occured in June will be

P_F_J = P_F_and_J / P_J

print(f'P(J) : {round(P_J*100,2)}%')
print(f'P(F AND J) : {round(P_F_and_J*100,2)}%')
print(f'P(F|J): {round(P_F_J*100,2)}%')
```

 #### <font color='orange'>Conclusion:</font>
 So, there is 73.33% chance of Floods when there is a heavy rainfall in Jun

Q2. Given that there is a flooding, calculate the probability that heavy rainfall has occurred in July (more than threshold value)?

Here we want to find  P(July=1|Flood=YES)

We are already aware of using formula based approach, so We will solve this using contingency table

```python
pd.crosstab(index = data['JUL'],
            columns = data['FLOODS'],
            margins=True,
            normalize='columns')
```

#### <font color='orange'>**Conclusion:**</font>  

The probability that high rainfall occurred in July (JUL = 1) given flooding (FLOODS = YES) is **0.65**.    

- **This means that when there is flooding, there is a 65% chance of heavy rainfall in July.**

### <font color='purple'>**Q3.Calculate the probability of flood given that june and july rainfall was greater than their median rainfall value?**</font>  

Solution:
We want to find  P(Flood=Yes| june=1 and Jul=1)

Here, we can pass multiple columns in the pd.crosstab()


```python
pd.crosstab(index = [data['JUN'], data['JUL']],
            columns = data['FLOODS'],
            margins=True,
            normalize='index')
```

### <font color='orange'>**Conclusion**</font>  

Frequency (JUN = 1, JUL = 1, FLOODS = YES ) = 0.9000000

There is **90%** chance of flood given that heavy rainfall in both june and july

```python
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px

# Load dataset
df = pd.read_csv("kerala.csv")


# Ensure correct column names (strip spaces)
df.columns = df.columns.str.strip()

# Convert FLOODS column to numeric (1 if YES, 0 otherwise)
df["FLOODS"] = df["FLOODS"].apply(lambda x: 1 if str(x).strip().upper() == "YES" else 0)

# Melt monthly rainfall data for trend plot
month_cols = [col for col in df.columns if col not in ["YEAR", "SUBDIVISION", "ANNUAL RAINFALL", "FLOODS"]]
df_melted = df.melt(id_vars=["YEAR", "SUBDIVISION"], value_vars=month_cols,
                    var_name="Month", value_name="Rainfall")

# Initialize app
app = dash.Dash(__name__)

# Layout
app.layout = html.Div([
    html.H1("Rainfall Dashboard", style={"textAlign": "center"}),

    dcc.Dropdown(
        id="subdivision-dropdown",
        options=[{"label": sub, "value": sub} for sub in df["SUBDIVISION"].unique()],
        value=df["SUBDIVISION"].unique()[0]
    ),

    html.Div(id="summary-cards", style={"display": "flex", "justifyContent": "space-around", "margin": "20px"}),

    dcc.Graph(id="monthly-trend"),
    dcc.Graph(id="annual-rainfall")
])

# Callback
@app.callback(
    [Output("summary-cards", "children"),
     Output("monthly-trend", "figure"),
     Output("annual-rainfall", "figure")],
    [Input("subdivision-dropdown", "value")]
)
def update_dashboard(selected_subdivision):
    df_filtered = df[df["SUBDIVISION"] == selected_subdivision]
    df_monthly = df_melted[df_melted["SUBDIVISION"] == selected_subdivision]

    # Summary stats
    avg_rain = round(df_filtered["ANNUAL RAINFALL"].mean(), 2)
    max_rain = df_filtered.loc[df_filtered["ANNUAL RAINFALL"].idxmax()]
    flood_years = df_filtered[df_filtered["FLOODS"] == 1]["YEAR"].tolist()

    summary_cards = [
        html.Div([
            html.H3("Avg Annual Rainfall"),
            html.P(f"{avg_rain} mm")
        ], style={"border": "1px solid black", "padding": "10px", "width": "30%"}),

        html.Div([
            html.H3("Max Rainfall Year"),
            html.P(f"{max_rain['YEAR']} - {max_rain['ANNUAL RAINFALL']} mm")
        ], style={"border": "1px solid black", "padding": "10px", "width": "30%"}),

        html.Div([
            html.H3("Flood Years Count"),
            html.P(f"{len(flood_years)} years")
        ], style={"border": "1px solid black", "padding": "10px", "width": "30%"})
    ]

    # Monthly rainfall trend
    fig_monthly = px.line(
        df_monthly.groupby("Month")["Rainfall"].mean().reset_index(),
        x="Month", y="Rainfall",
        title="Average Monthly Rainfall Trend"
    )

    # Annual rainfall with floods highlighted
    fig_annual = px.bar(
        df_filtered, x="YEAR", y="ANNUAL RAINFALL", color="FLOODS",
        title="Annual Rainfall & Flood Years",
        color_discrete_map={0: "blue", 1: "red"}
    )

    return summary_cards, fig_monthly, fig_annual

# Run app
if __name__ == "__main__":
    app.run(debug=True)
```
